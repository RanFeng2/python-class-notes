# 5.3  CSV文件操作

Comma Separated Values，简称CSV，它是一种以逗号分隔数值的文件类型。在数据库或电子表格中，它是最常见的导入导出格式，它以一种简单而明了的方式存储和共享数据，CSV文件通常以纯文本的方式存储数据表。

两种方法：`reader`,`writer`

   `reader(iterable [, dialect='excel'], [optional keyword args])`

- `reader`函数用来读取csv格式的文件内容，并将结果保存到一个迭代器中，`iterable`表示读取的**文件对象**或**列表对象**，dialect表示默认的格式为兼容“excel”的格式。


`writer(fileobj [, dialect='excel'], [optional keyword args])`

- `writer`函数用来写入csv格式文件，其中`fileobj`为需要写入的**文件对象**。

### 2.读取文件某一列数据

```python
#1.获取文件某一列数据
import csv
with open("test1.csv") as f:
    reader = csv.reader(f)
    column=[row[0] for row in  reader]
    print(column)
```

### 3.向csv文件中写入数据

```python
#1.向csv文件中写入数据,a-->append
import csv
with open("test1.csv",'a') as f:
    row=['曹操','23','学生','黑龙江','5000']
    write=csv.writer(f)
    write.writerow(row)
    print("写入完毕！")
```

### 4.获取文件头及其索引

```python
import csv
with open("test1.csv") as f:
    #1.创建阅读器对象
    reader = csv.reader(f)
    #2.读取文件第一行数据
    head_row=next(reader)
    print(head_row)
    #4.获取文件头及其索引
    for index,column_header in enumerate(head_row):
        print(index,column_header)
```

```python
head_row.index("姓")
```

### 5.获取某列的最大值

```python
# ['姓名', '年龄', '职业', '家庭地址', '工资']
import csv
with open("test1.csv") as f:
    reader = csv.reader(f)
    header_row=next(reader)
    # print(header_row)
    salary=[]
    for row in reader:
        #把第五列数据保存到列表salary中
         salary.append(int(row[4]))
    print(salary)
    print("员工最高工资为："+str(max(salary)))
```

### 6.复制CSV格式文件

```python
import csv
f=open('test1.csv')
#1.newline=''消除空格行
aim_file=open('Aim.csv','w',newline='')
write=csv.writer(aim_file)
reader=csv.reader(f)
rows=[row for row in reader]
#2.遍历rows列表
for row in rows:
    #3.把每一行写到Aim.csv中
    write.writerow(row)
```

```python
f.close()
aim_file.close()
```

### 7. csv文件转换为json文件

csv转换为包含字典列表：
    key:表头
    values：表头下面的内容


```python
filename = "test1.csv"
with open(filename, 'r') as f:
    r = csv.reader(f)
    test_list= [eachrow for eachrow in r]
test_list
```

```python
keys = test_list[0]
values = test_list[1:]
values
```

```python
dict(zip(keys,values[0]))
```

```python
test_dict = [dict(zip(keys,eachvalue)) for eachvalue in values]
```

```python
test_dict
```

```python
import json
with open('test2020.json', 'w') as fp:
    json.dump(test_dict, fp, indent=4, separators=[',', ':'])
```
