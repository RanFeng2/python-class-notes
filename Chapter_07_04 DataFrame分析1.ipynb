#### 对行进行切片，注意切片是限定的左闭右开区间

`a['x']` 那么将会返回columns为x的列，注意这种方式一次只能返回一个列。  
`a.x`与`a['x']`意思一样。

取行数据，通过切片`[]`来选择  
如：`a[0:3]` 则会返回前三行的数据。

````python
#下标在[5,10]区间的的行
df[5:11]
````

#### loc和at使用标签文本做索引

- loc是通过标签来选择数据    
`a.loc['one']`则会默认表示选取行为'one'的行；  
`a.loc[:,['a','b'] ] `表示选取所有的行以及columns为a,b的列；  
`a.loc[['one','two'],['a','b']] `表示选取'one'和'two'这两行以及columns为a,b的列；
`a.loc['one','a']`与`a.loc[['one'],['a']]`作用是一样的，不过前者只显示对应的值，而后者会显示对应的行和列标签。

#### iloc使用数字做索引,iat()

- iloc则是直接通过位置来选择数据  
`a.iloc[1:2,1:2] `则会显示第一行第一列的数据;(切片后面的值取不到)  
`a.iloc[1:2]` 即后面表示列的值没有时，默认选取行位置为1的数据;  
`a.iloc[[0,2],[1,2]]` 即可以自由选取行位置，和列位置对应的数据。

````python
# iloc使用数字做索引
# 索引为5的行
df.iloc[5]
````

````python
# 下标为[3,5,10]的行
df.iloc[[0,5,10],:]
````

````python
# 行下标为[3,5,10]，列下标为[0,1,4]
df.iloc[[3,5,10],[0,1,4]]
````

````python
# 查看指定的列前5行数据
df[['工号', '时段', '交易额']][:5]
````

````python
#只查看前10行指定的列
df[:10][['工号', '日期', '柜台']]
````

````python
# 下标为[3,5,10]的行的指定列
df.loc[[3,5,10], ['姓名','交易额']]
````

````python
# 行下标为3,姓名列的值
df.at[3, '姓名']
````

````python
df.iat[3,1]
````

## 5 分组与聚合

### 分组 groupby

````python
import pandas as pd
import numpy as np
# 读取全部数据，使用默认索引
df = pd.read_excel(r'超市营业额2.xlsx')
df[:10]
````

````python
# 根据lambda表达式对index处理后的结果分组'.ljust(30,'='))
df.groupby(by=lambda num: num%5)['交易额'].sum()
````

````python
# 根据指定字典的“键”对index进行分组，“值”作为index标签
# 指定by参数为字典'.ljust(30,'='))
df.groupby(by={7:'下标为7的行', 35:'下标为35的行'})['交易额'].sum()
````

````python
# 不同时段的销售总额'.ljust(30,'='))
df.groupby(by='时段')['交易额'].sum()
````

````python
# 各柜台的销售总额'.ljust(30,'='))
df.groupby(by='柜台')['交易额'].sum()
````

````python
# 可以查看每个员工上班总时长是否均匀
# 每个员工上班的次数'.ljust(30,'='))
dff = df.groupby(by='姓名')['日期'].count()
dff.name = '上班次数'
dff
````

````python
# 每个员工交易额平均值'.ljust(30,'='))
df.groupby(by='姓名')[ '交易额'].mean().round(2).sort_values()
````

````python
# 汇总交易额转换为整数'.ljust(30,'='))
df.groupby(by='姓名').sum()['交易额'].apply(int)
````

````python
# 每个员工交易额的中值'.ljust(30,'='))
dff = df.groupby(by='姓名').median()
dff['交易额']
````

````python
dff['排名'] = dff['交易额'].rank(ascending=False)
# 每个员工交易额中值的排名'.ljust(30,'='))
dff[['交易额','排名']]
````

````python
# 每个员工不同时段的交易额'.ljust(30,'='))
df.groupby(by=['姓名','时段'])['交易额'].sum()
````

### 聚合aggregate()

````python
# 对不同的列可以采用不同的函数
# 时段和交易额采用不同的聚合方式
df.groupby(by=['姓名'])['时段', '交易额'].aggregate({'交易额':np.sum, '时段':lambda x:'各时段累计'})
````

````python
df
````

````python
# 使用DataFrame结构的agg()方法对指定列进行聚合
# 使用agg()方法对交易额进行聚合'.ljust(30,'='))
df.agg({'交易额':['sum','mean','min','max','median'], '日期':['min','max']})
````

````python
# 对分组结果进行聚合'.ljust(30,'='))
df.groupby(by='姓名').agg(['max','min','mean','median'])
````

````python
# 查看分组聚合后的部分结果'.ljust(30,'='))
df.groupby(by='姓名').agg(['max','min', 'mean','median'])['交易额']
````

## 6  处理异常值

````python
import pandas as pd

# 读取全部数据，使用默认索引
df = pd.read_excel(r'超市营业额2.xlsx')

# 查看交易额低于200的数据'.ljust(20,'='))
df[df.交易额<200]
````

````python
df.loc[df.交易额<200, '交易额'] = df[df.交易额<200]['交易额'].map(lambda num: num*1.5)
print('上浮50%之后仍低于200的数据'.ljust(20,'='))
df[df.交易额<200]
````

````python
# 查看交易额高于3000的数据'.ljust(20,'='))
df[df['交易额']>3000]
````

````python
# 交易额低于200或高于3000的数据'.ljust(20,'='))
df[(df.交易额<200) | (df.交易额>3000)]
````

````python
# 把低于200的交易额都替换为固定的200
df.loc[df.交易额<200, '交易额'] = 200
print('交易额低于200或高于3000的数据'.ljust(20,'='))
df[(df.交易额<200) | (df.交易额>3000)]
````

````python
# 把高于3000的交易额都替换为固定的3000
df.loc[df.交易额>3000, '交易额'] = 3000
print('交易额低于200或高于3000的数量'.ljust(20,'='))
df[(df.交易额<200) | (df.交易额>3000)]['交易额'].count()
````

## 7 处理超市交易数据中的缺失值

````python
from copy import deepcopy
import pandas as pd

# 读取全部数据，使用默认索引
df = pd.read_excel(r'超市营业额2.xlsx')
````

````python
# 数据总行数'.ljust(20,'='))
len(df)
````

````python
# 丢弃缺失值之后的行数'.ljust(20,'='))
len(df.dropna())
````

````python
# 包含缺失值的行'.ljust(20,'='))
df[df['交易额'].isnull()]
````

````python
df
````

````python
# 使用固定值替换缺失值'.ljust(20,'='))
# 深复制，不影响原来的df,[[1,2],12,13]
import copy
dff = copy.deepcopy(df)
````

````python
dff.loc[dff.交易额.isnull(),'交易额'] = 1000
dff.iloc[[110,124,168],:]
````

````python
# 使用每人交易额均值替换缺失值'.ljust(20,'='))
dff = copy.deepcopy(df)
for i in dff[dff.交易额.isnull()].index:
    dff.loc[i, '交易额'] = round(dff.loc[dff.姓名==dff.loc[i,'姓名'], '交易额'].mean())
dff.iloc[[110,124,168],:]
````

````python
# 使用整体均值的80%填充缺失值'.ljust(20,'='))
df.fillna({'交易额': round(df['交易额'].mean()*0.8)}, inplace=True)
df.iloc[[110,124,168],:]
````

## 8  处理重复值

````python
from copy import deepcopy
import pandas as pd
import numpy as np

# 读取全部数据，使用默认索引
df = pd.read_excel(r'超市营业额2.xlsx')

# 数据总行数'.ljust(20,'='))
len(df)
````

````python
# 重复行'.ljust(20,'='))
df[df.duplicated()]
````

````python
# 一人同时负责两个柜台的排班'.ljust(20,'='))
dff = df[['工号','姓名','日期','时段']]
dff = dff[dff.duplicated()]
for row in dff.values:
    print(df[(df.工号==row[0])&(df.日期==row[2])&(df.时段==row[3])])
````

````python
# 直接丢弃重复行
df = df.drop_duplicates()
len(df)
````

````python
# 可以查看是否有录入错误的工号和姓名
# 所有工号与姓名的对应关系'.ljust(20,'='))
dff = df[['工号','姓名']]
dff.drop_duplicates()
````

## 9 使用数据差分查看员工业绩波动情况

````python
import pandas as pd

# 读取全部数据，使用默认索引
df = pd.read_excel(r'超市营业额2.xlsx')
````

````python
# 每天交易总额变化情况'.ljust(20,'='))
dff = df.groupby(by='日期').sum()['交易额'].diff()
# 格式化，整数前面带加号
dff.map(lambda num:'%+.2f'%num)[:5]
````

````python
# 张三的每天交易总额变化情况'.ljust(20,'='))
df[df.姓名=='张三'].groupby(by='日期').sum()['交易额'].diff()[:5]
````
