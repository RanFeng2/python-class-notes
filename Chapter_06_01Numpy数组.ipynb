### 6.0.2 Numpy的主要对象是**同质的多维数组**。

  其中的元素通常都是**数字**，并且是同样的类型，由一个正整数元组进行索引。每个元素在内存中占有同样大小的空间。

- 在Numpy中，维度被称为‘轴’。

> 例如对于[1, 2, 1]我们说它有一个轴，并且长度为3。而[[ 1., 0., 0.], [ 0., 1., 2.]]则有两个轴，第一个轴的长度为2，第二个轴的长度为3。

- Numpy数组类的名字叫做`ndarray`.

- `np.array`的代码定义如下：

     `numpy.array(object, dtype = None, copy = True, order = None, subok = False, ndmin = 0)`

各参数的含义：

`object`：用于生成数组的数据对象<br>
`dtype`：指定类型，可选。<br>
`copy`：可选，默认为`True`，对象被复制。<br>
`order`：C语言风格（按行）、FORTRAN风格（按列）或A（任意，默认）。<br>
`subok`：默认情况下，返回的数组被强制为基类数组。 如果为`Tru`e，则返回子类。<br>
`ndmin`：指定返回数组的最小维数。<br>

### 6.0.3 `ndarray`具有以下重要属性：

`ndarray.ndim`：数组的轴数量<br>
`ndarray.shape`：数组的形状。这是一个整数元组。比如对于n行m列的矩阵，其shape形状就是(n,m)。而shape元组的长度则恰恰是上面的ndim值，也就是轴数。<br>
`ndarray.size`：数组中所有元素的个数。这恰好等于shape中元素的乘积。<br>
`ndarray.dtype`：数组中元素的数据类型。除了标准的Python类型，Numpy还提供一些自有的类型。<br>
`ndarray.itemsize`：元素的字节大小。比如float64类型的itemsize为8（=64/8），而complex32的itemsize为4（=32/8）。<br>
`ndarray.data`：包含数组实际元素的缓冲区。通常我们不需要使用这个属性，因为我们将使用索引工具访问数组中的元素。<br>
`ndarray.flags`: 数组对象的一些状态指示或标签

##### (2) linspace()

````python
from numpy import pi  # 导入圆周率
np.linspace( 0, 2, 9 )                 # 从0到2之间的9个数
````

````python
x = np.linspace( 0, 2*pi, 100 )        # 从0到2Π之间，生成100个数
f = np.sin(x)
f
````

- 其原型为：

`numpy.linspace(start, stop, num, endpoint, retstep, dtype)`

`start`： 序列的起始值<br>
`stop`： 序列的终止值，如果`endpoint`为`True`，则终止值包含于序列中<br>
`num`： 要生成的等间隔样例数量，默认为`50`<br>
`endpoint`： 序列中是否包含stop值，默认为`Ture`<br>
`retstep`： 如果为`True`，返回样例以及连续数字之间的步长<br>
`dtype`： 输出`ndarray`的数据类型<br>

## 6.1.2  测试两个数组的对应元素是否足够接近

`isclose()`,`allclose()`函数来测试两个数组中对应位置上的元素在允许的范围内是否相等。并可以接收绝对误差参数和相对误差参数。
- `isclose()`用来测试每一对元素是否相等并返回若干`True/False`的列表
- `allclose()`用来测试所有对应位置上的元素是否相等，并返回单个`True`或`False`。

````python
import numpy as np
x = np.array([1, 2, 3, 4.001, 5])
y = np.array([1, 1.999, 3, 4.01, 5.1])
````

````python
print(np.isclose(x, y))
````

````python
print(np.allclose(x, y, rtol=0.2))       # 设置相对误差参数
print(np.allclose(x, y, atol=0.2))       # 设置绝对误差参数
````

````python
print(np.isclose(x, y))
print(np.isclose(x, y, atol=0.2))
````

#### 6 numpy中赋值、视图、深复制

##### （1）赋值

当对numpy数组进行赋值时，只是对同一个对象新建了一个引用，并不是建立新的对象，所以赋值前后的变量完全是同一对象，对其中一个引用修改时，所有引用都会生效：

a = np.arange(12)
b = a # 赋值
a is b

b.shape = (3, 4)
b
a


##### （2）视图（切片、浅复制）

numpy中允许不同数组间共享数据，这种机制在numpy中称为视图，对numpy数组的切片和浅复制都是通过视图实现的。如果数组B是数组A的视图(view)，则称A为B的base(除非A也是视图)。视图数组中的数据实际上保存在base数组中。

````python
a = np.arange(12)
b = a.view() # 使用视图
a is b
````

````python
b
````

````python
b.shape = (3, 4) # 改变b的形状
a
````

````python
b
````

````python
b[0] = 0
a
````

````python
b
````

从上面代码中我们可以发现，a和b是不同的两个数组，改变b的形状对a不会有影响，但是改变b的数据之后，a的数据也会发现改变，说明a与b是共享数据的。

再来探索一些切片：

````python
a = np.arange(12)
b = a[:] # 切片
a is b
````

````python
b.shape = (3, 4)
a
````

````python
b
````

````python
b[0] = 0
a
````

````python
b
````

果然，切片效果与视图一致。

##### （3）深复制

深复制通过数组自带的copy()方法实现，深复制产生的数组与原数组时完全不同的两个数组对象，完全享有独立的内存空间，所有操作都不会相互影响。

````python
a = np.arange(12)
b = a.copy()
a is b

b.shape = (3, 4)
b[0] = 0
a

b
````
