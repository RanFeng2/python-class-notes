```python
import json
dicc = {'Tom':81,'Jerry':95}
dic = json.dumps(dicc)
print(dic,type(dic))
dic
```

```python
import json
with open ('test.json','w',encoding='utf-8') as f:
    json.dump(dicc,f,indent=4)  #indent 缩进，每一个层级向里面缩进四个格子
```

```python
import json
with open('test.json','r',encoding='utf-8') as f:
    #data = json.loads(f.read())
    data = json.load(f)
    print(data,type(data))
    f.seek(0)
    data1 = json.loads(f.read())
    print(data1,type(data1))
```

```python
import json
with open ('房屋信息.json','r',encoding='gbk') as fp:
    info = json.load(fp)
    print(info)
```

```python
import json 
with open ('房屋信息复制版.json','w') as fp:
    json.dump(info,fp,indent=4,separators=[',',':'])
```

```python
import json
with open ('房屋信息复制版.json','r') as fp:
    infoo = json.load(fp)
    for i in infoo:
        print(i)
```

```python
with open ('房子.txt','r',encoding='utf-8') as fp:
    da = fp.readlines()
da
```

```python
daa = [eval(each) for each in da]
daa
```

```python
import json
with open('房子.json','w') as f:
    json.dump(daa,f,indent=4,separators=[',',':'])
```

```python
import json
with open ('房子.json','r') as ff:
    inda = json.load(ff)
for i in inda:
    print(i)
```
