# 5.2  `JSON`文件操作

- `JSON (JavaScript Object Notation)` 是一种轻量级的数据交换格式。`Python3` 中可以使用 `json` 模块来对 `JSON` 数据进行编解码。

- json.dumps 将 Python 对象编码成 JSON **字符串**
- json.loads 将已编码的 JSON 字符串解码为 Python 对象
- json.dump和json.load,需要传入文件描述符，加上文件操作。
- JSON内部的格式要注意，一个好的格式能够方便读取，可以用indent格式化
> indent: 设置缩进格式，默认值为None,选择的是最紧凑的表示。如果indent是非负整数或字符串，那么JSON数组元素和对象成员将使用该缩进级别进行输入；indent为0,负数或“”仅插入换行符；indent使用正整数缩进多个空格；如果indent是一个字符串（例如“\t”），则该字符串用于缩进每个级别。

JSON建构于两种结构： 
- “名称/值”对的集合：在Python语言中，它被理解为字典；
-  值的有序列表：在Python语言中，它被理解为列表。


### dumps

将一个Python对象进行JSON格式的编码。


| Python                                 | JSON   |
| -------------------------------------- | ------ |
| dict                                   | object |
| list, tuple                            | array  |
| str                                    | string |
| int, float, int- & float-derived Enums | number |
| True                                   | true   |
| False                                  | false  |
| None                                   | null   |

```python
import json

# dumps可以格式化所有的基本数据类型为字符串
data1 = json.dumps([])         # 列表
print(data1, type(data1))
data1
```

```python
data2 = json.dumps(2)          # 数字
print(data2, type(data2))
data2
```

```python
data3 = json.dumps('3')        # 字符串
print(data3, type(data3))
```

```python
dict = {"name": "Tom", "age": 23}   # 字典
data4 = json.dumps(dict)
#dict
data4
print(data4, type(data4))
```

```python
with open("test.json", "w", encoding='utf-8') as f:
    # indent 超级好用，格式化保存字典，默认为None，小于0为零个空格
#    f.write(json.dumps(dict, indent=4))
    json.dump(dict, f, indent=4)  # 传入文件描述符，和dumps一样的结果
```

### loads

loads将json格式数据解码为python对象。

```python
import json

dict = '{"name": "Tom", "age": 23}'   # 将字符串还原为dict,eval
data1 = json.loads(dict)
print(data1, type(data1))
data1
```

```python
import json
with open("test.json", "r", encoding='utf-8') as f:
    data2 = json.loads(f.read())    # load的传入参数为字符串类型
    print(data2, type(data2))
    f.seek(0)                       # 将文件游标移动到文件开头位置
    data3 = json.load(f)
    print(data3, type(data3))
```

### JSON中的数据格式和Python中的数据格式转化关系如下：

| JSON          | Python |
| ------------- | ------ |
| object        | dict   |
| array         | list   |
| string        | str    |
| number (int)  | int    |
| number (real) | float  |
| true          | True   |
| false         | False  |
| null          | None   |

### 例5-2  
把包含若干房屋信息的列表写入`JSON`文件，然后再读取并输出这些信息。

```python
import json

information = [
               {'小区名称': '小区A', '均价': 8000, '月交易量': 20},
               {'小区名称': '小区B', '均价': 8500, '月交易量': 35},
               {'小区名称': '小区C', '均价': 7800, '月交易量': 50},
               {'小区名称': '小区D', '均价': 12000, '月交易量': 18}]

with open('房屋信息.json', 'w') as fp:
    json.dump(information, fp, indent=4, separators=[',', ':'])
```

```python
with open('房屋信息.json') as fp:
    information = json.load(fp)
    for info in information:
        print(info)
```

```python
with open('房屋信息.json') as fp:
    information = json.load(fp)

print(information)
```
