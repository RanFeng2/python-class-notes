### (1) pd.timeStamp()时间戳

```python
date1 = "2019-05-01 12:25:12"
date2 = datetime.datetime(2019,5,6,14,15,23)
t1 = pd.Timestamp(date1)  #时间戳
t2 = pd.Timestamp(date2)
print(t1,type(t1))
print(t2,type(t2))
print(date2,type(date2))
```

### （2）pd.to_datetime()

```python
date1 = "2017-05-01 12:25:12"
date2 = datetime.datetime(2017,5,6,14,15,23)
 
t1 = pd.to_datetime(date1)  #时间戳
t2 = pd.to_datetime(date2)
print(t1,type(t1))
print(t2,type(t2))
```

```python
# 多个时间数据，会转化成pandas的Datetime的Index
ls_date = ['2017-01-01','2017-01-02','2017-01-03']
t3 = pd.to_datetime(ls_date)
print(t3,type(t3))
```

```python
#当一组数据中夹杂着其他的数组
date3 = ['2017-01-01','2017-01-02','2017-01-03','hello','2018-01-05']
t4 = pd.to_datetime(date3,errors='ignore')  #返回原始数据，这里直接是生成一组数据
print(t4,type(t4))
```

```python
t5 = pd.to_datetime(date3,errors='coerce')  #缺失值返回NaT，结果是DatetimeIndex
print(t5,type(t5))
```

### (3) date_range(start,end,periods,freq)

时间序列对象一般使用pandas的date_range()函数生成，可以指定日期时间的起始和结束范围、时间间隔以及数据数量等参数，语法为：

- start指定起始日期，end指定结束日期
- periods指定产生的数据数量
- freq指定间隔，
    - D表示天，W表示周，H表示小时
    - M表示月末最后一天，MS表示月初第一天
    - T表示分钟，Y表示年末最后一天，YS表示年初第一天

```python
import pandas as pd
# 间隔5天
pd.date_range(start='20190601', end='20190630', freq='5D')
```

```python
# 间隔1周
pd.date_range(start='20190601', end='20190630', freq='W')
```

```python
# 间隔2天，5个数据
pd.date_range(start='20190601', periods=5, freq='2D')
```

```python
#间隔3小时，8个数据
pd.date_range(start='20190601', periods=8, freq='3H')
```

```python
# 3:00开始，间隔1分钟，12个数据
print(pd.date_range(start='201906010300', periods=12, freq='T'))
```

```python
# 间隔1月，月末最后一天
pd.date_range(start='20190101', end='20191231', freq='MS')
```

```python
# 间隔1年，6个数据，年末最后一天
pd.date_range(start='20190101', periods=6, freq='A')
```

```python
# 间隔1年，6个数据，年初第一天
pd.date_range(start='20190101', periods=6, freq='AS')
```

```python
#复合频率
print(pd.date_range('20180701','20180801',freq='7D'))#7天
print(pd.date_range('20180701','20180801',freq='2H30min'))#2小时30分钟
print(pd.date_range('2018','2019',freq='2MS'))       #2月，每月最后一个日历日
```

##### 使用日期时间做索引，创建Series对象

```python
data = pd.Series(index=pd.date_range(start='20190701', periods=24, freq='H'), data=range(24))
print('前5条数据'.ljust(30, '='))
print(data[:5])
```

### (4) resample()重采样函数 

```python
# 3分钟重采样，计算均值
data.resample('3H').mean()
```

```python
# 5小时重采样，求和
print(data.resample('5H').sum())
```

```python
# OHLC分别表示OPEN、HIGH、LOW、CLOSE
# 5小时重采样，统计OHLC值
data.resample('5H').ohlc()
```
