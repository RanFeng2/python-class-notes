## 6.1.11  数组布尔运算


使用布尔数组进行索引，其实就是我们显式地选择数组中需要哪些项，不需要哪些项。

### 1 使用与原始数组形状相同的布尔数组进行筛选过滤：

```python
a = np.arange(12).reshape(3,4)
a
```

```python
b = a > 4
b   # 通过比较运算，b变成了一个由布尔值组成的数组
```

```python
a[a>4]        # 生成一个由True值对应出来的一维数组
```

```python
a
```

这个技巧用在特殊位置赋值操作特别高效：

```python
a[b] = 0    #所有a中大于4的元素被重新赋值为0
a
```

```python
a[(a>4) & (a<8)]
```

实际上，上面的操作可以简写成：a[a>4] = 0

> 使用~可以对布尔值取反，|表示或，&表示与：

```python
a[~b]
```

```python
(a<4)|(a>7)
```

```python
a[(a>3)&(a<8)]
```

```python
(a>3)&(a<8)
```

## 6.1.12  分段函数

numpy提供了`where()`和`piecewise()`两个函数支持分段函数对数组的处理，

### 1 `where()`

适合对原数组进行“二值化”，根据数组中的元素是否满足指定的条件返回x还是y。语法如下：
      
      where(condition,[x,y])
      

```python
import numpy as np
x = np.random.randint(0, 10, size=(1,10))
x
```

```python
np.where(x<5, 0, 1)          # 小于5的元素值对应0，其他对应1
```

### 2 `piecewise()`

可以实现更复杂的处理，语法如下：

      piecewise(x,condlist,funclist,*args,**kw)

```python
x
```

```python
np.piecewise(x, [x<4, x>7], [lambda x:x*2, lambda x:x*3])
 # 小于4的元素乘以2，大于7的元素乘以3，其他元素变为0
```

```python
np.piecewise(x, [x<3, (3<x)&(x<5), x>7], [-1, 1, lambda x:x*4])
# 小于3的元素变为-1，大于3小于5的元素变为1，大于7的元素乘以4
# 条件没有覆盖到的其他元素变为0
```
