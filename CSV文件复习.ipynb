## 复制csv格式文件

```python
import csv
with open ('test1.csv') as fp1,open ('Aim.csv','w',newline='') as fp2:
    reader = csv.reader(fp1)
    write = csv.writer(fp2)
    rows = [row for row in reader]
    for row in rows:
        write.writerow(row)
```
