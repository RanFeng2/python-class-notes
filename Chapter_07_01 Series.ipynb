## 7.0 引言

#### Pandas是一个高性能的数据操作和分析工具。

### 4 Series的运算

> 注意：在操作Series的时候，基本可以把Series看成NumPy中的ndarray数据来进行操作。


#### （1）两个等长Series对象之间可以进行四则运算和幂运算
- 只对两个Series对象中都有的索引对应的值进行计算
- 非共同索引对应的值为控制NaN

```python
# 两个Series对象相加
print(pd.Series(range(5))+pd.Series(range(5,10)))
```

```python
# s2与数字之间的运算
round((s2**0.5)*10, 1)
```

```python
#5.Series值的运算
#Series中元素级别的运算结果，包含索引值并且键值关系保持不变
print("=========Series值的运算==========")
s6 = pd.Series({'a':1,'b':2,"c":3,"d":4})
print(s6)
print("=========打印Series大于2的值==========")
print(s6[s6>2])
print("=========打印Series的值除以2==========")
print(s6/2)
```

```python
#numpy中的通用函数在Series中也支持
s7= pd.Series([1,2,-3,-4])
print(np.exp(s7))
```

#### （2） pipe()方法可以实现函数链式调用的功能

```python
#每个值的平方对5的余数
pd.Series(range(5)).pipe(lambda x,y,z:(x**y)%z, 2, 5)
```

```python
# 每个值加3之后再乘以3
pd.Series(range(5)).pipe(lambda x:x+3).pipe(lambda x:x*3)
```

#### （3）apply()方法

```python
# 每个值加3
pd.Series(range(5)).apply(lambda x: x+3)
```

#### （4）标准差、无偏方差、无偏标准差

```python
pd.Series(range(5)).std()
pd.Series(range(5)).var()
pd.Series(range(5)).sem()
```

#### （5）any(),all()
- any()查看是否存在等价于True的值

- all()查看是否所有值都等价于True

```python
any(pd.Series([3,0,True]))
```

```python
all(pd.Series([3,0,True]))
```

#### （6）Series自动对齐

当多个series对象之间进行运算的时候，如果不同series之间具有不同的索引值，那么运算会自动对齐不同索引值的数据，如果某个series没有某个索引值，那么最终结果会赋值为NaN。

```python
# 7.Series自动对齐
 
s8 = pd.Series([12,28,46],index=["p1","p2","p3"])
s9 = pd.Series([2,4,6,8],index=["p2","p3","p4","p5"])
print("=======s8=======")
print(s8)
print("=======s9=======")
print(s9)
```

```python
print("=======s8+s9=======")
s10 = s8+s9
s10
```


#### 在Pandas中，可以使用isnull和notnull函数来检查缺失的数据：

```python
pd.isnull(s10)
```
