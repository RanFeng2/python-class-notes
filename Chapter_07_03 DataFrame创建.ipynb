## 7.3 DataFrame

- DataFrame是Pandas的核心数据结构，表示的是**二维的矩阵数据表**，类似关系型数据库的结构，每一列可以是不同的值类型，比如数值、字符串、布尔值等等。<br>
- 可以将其视为电子表格或SQL表。<br>
- DataFrame既有**行索引(index)**， 也有 **列索引(columns)**，还有**值（values）**它可以被看做为一个共享相同索引的Series的字典。

## 1创建DataFrame对象

###  (1) 由list创建

```python
import numpy as np
import pandas as pd
```

```python
#由list和数组创建
data = {
    'name':['Jack','Mary','Tom'],
    'age':[14,15,17],
    'gender':['M','W','M']
}
fr = pd.DataFrame(data)
fr
```

```python
print(type(fr))
print(fr.index,'数据类型是：',type(fr.index))   #行标签
print(fr.values,'数据类型是：',type(fr.values)) #值
print(fr.columns,'数据类型是：',type(fr.columns))  #列标签
```

### (2) 由Series组成的创建

```python
#由Series组成的创建
data1 = {'one':pd.Series(np.random.rand(2)),
         'two':pd.Series(np.random.rand(3)),
         }
data1
```

```python
fr1 = pd.DataFrame(data1)
fr1
```

```python
data2 = {'one':pd.Series(np.random.rand(2),index=['a','b']),
         'two':pd.Series(np.random.rand(3),index=['a','b','c'])
         }
```

```python
fr2 = pd.DataFrame(data2)
fr2
```

### (3) 通过二维数组创建 （常用）

```python
#通过二维数组创建 （常用）
ar = np.random.rand(9).reshape(3,3)
ar
```

```python
fr3 = pd.DataFrame(ar)
fr3
```

```python
fr4 = pd.DataFrame(ar,index=['a','b','c'],columns=['s','h','j'])
fr4
```

### (4) 字典组成的列表

```python
data3 = [{'one':1,'two':2,'three':3},{'four':4,'five':5,'six':6}]
data3
```

```python
fr5 = pd.DataFrame(data3)
fr5
```

### (5) 字典组成的字典

```python
data4 = {
    'Tom':{'art':67,'english':98,'china':76},
    'Mary':{'art':45,'english':78,'china':70},
    'Lucy':{'art':58,'english':79},
        }
fr6 = pd.DataFrame(data4)
fr6
```

## 2 DataFrame查看

#### （1）head(),tail()函数
可以使用head方法查看DataFrame对象的前5行，用tail方法查看后5行。或者head(3)，tail(3)指定查看行数：

```python
data1 = {'state':['beijing','beijing','beijing','shanghai','shanghai','shanghai'],
 'year':[2000,2001,2002,2001,2002,2003],
 'pop':[1.5, 1.7,3.6,2.4,2.9,3.2]}
df4 = pd.DataFrame(data1)
df4
```

```python
df4.head(2)
```

```python
df4.tail(3)
```

#### (2)列排序
DataFrame对象中的state/year/pop，其实就是列索引，可以在创建的时候使用参数名columns指定它们的先后顺序：

```python
pd.DataFrame(data1, columns=['year','state','pop'])
```

#### （3）定行索引

当然，也可以使用参数名index指定行索引：

```python
df = pd.DataFrame(np.random.rand(12).reshape(3,4)*100,
                  index=['one','two','three'],columns=['a','b','c','d']
                  )
df2 = pd.DataFrame(np.random.rand(12).reshape(3,4)*100,
                  columns=['a','b','c','d']
                  )
 
print(df)
print(df2)
```

```python
data = df['a']
data1 = df[['a','b']]
print('data',data)
print('data1',data1)     #选择列
```

```python
data3 = df.loc['one']    #按index选择行,iloc
data4 = df.loc[['three','one']]
print('data3',data3)
print('data4',data4)   #选择行
```

```python
data5 = df.iloc[-1]  #按整数位置选择行
print('data5',data5) 
```

```python
print("单标签索引")
print(df.loc['one'])
```

```python
df2.loc[1]
```

```python
df
```

```python
df.loc[['two','one']]
```

```python
df2
```

```python
df2.loc[[2,1]]
```

```python
print("切片索引/n")
print(df.loc['one':'two'])
print(df2.loc[1:2])
```

#### (4)使用columns列索引来检索一列：

```python
data1 = {'state':['beijing','beijing','beijing','shanghai','shanghai','shanghai'],
 'year':[2000,2001,2002,2001,2002,2003],
 'pop':[1.5, 1.7,3.6,2.4,2.9,3.2]}
df5 = pd.DataFrame(data1, columns=['year','state','pop'],index=['a','b','c','d','e','f'])
df5
```

## 3 DataFrame修改

### （1）DataFrame对象追加列

```python
df5['debt'] = 12
df5
```

```python
df5['debt'] = np.arange(1,7)
df5
```

```python
val = pd.Series([1,2,3],index = ['c','d','f'])
df5['debt'] = val
df5  # 缺失值以NaN填补
```

#### （2）给DataFrame追加行

```python
data = {'state':['beijing','beijing','beijing','shanghai','shanghai','shanghai'],
'year':[2000,2001,2002,2001,2002,2003],
'pop':[1.5, 1.7,3.6,2.4,2.9,3.2]}

df = pd.DataFrame(data,index=list('abcdef'))
df
```

```python
df1 = df.loc['a']
```

```python
df.append(df1)
```

```python
df
```

#### （3）使用del方法删除指定的列：

```python
df['new'] = df.state=='beijing'
df
```

```python
df 
# 要注意的是我们有时候不能这么调用df的某个列，并执行某个操作。这是个坑。
```

```python
del df['new']
```

```python
df
```

> 需要注意的是：从DataFrame中选取的列是数据的视图，而不是拷贝。因此，对选取列的修改会反映到DataFrame上。如果需要复制，应当使用copy方法。
